## Revisión de inconsistencias en los datos

Aunque ya se ha realizado una exploración de los registros duplicados, es necesario investigar si los datos cumplen con el requisito de tener máximo 9 asignaturas cargadas en un mismo periodo (primavera, verano, otoño e invierno). Para lograrlo, se pueden agrupar los datos por `periodo` y `n_matricula` con la finalidad de realizar un conteo y filtrar aquellas observaciones que tengan más de 9 asignaturas por periodo. 

### Datos del 2016

```python
def asignaturas_por_periodo(dataframe):
    """
    Permite obtener la cantidad de asignaturas por
    semestre agrupadas por periodo y matricula. 
    Devuelve el resultado como dataframe.
    """
    # Agrupamos datos por periodo y matricula para obtener la cantidad de asignaturas por grupo
    materias_por_semestre = dataframe.groupby(["periodo", "n_matricula"]).size()
    # Obtenemos las materias por semestre en un dataframe ordenadas por cantidad de asignaturas y matriculas de mayor a menor
    materias_por_semestre = materias_por_semestre.reset_index(name = "n_asignaturas").sort_values(by = ["n_asignaturas", "n_matricula"], ascending = False)
    return(materias_por_semestre)
```

```python
# Calculamos las materias por semestre para este año
materias_por_semestre = asignaturas_por_periodo(datos["2016"])
# Mostramos el dataframe
materias_por_semestre
```

Como pudimos notar en la salida anterior, existen estudiantes (que se presentan con una matricula) que exceden el máximo permitido de asignaturas cargadas en un semestre. Es necesario 
explorar detalladamente estos registros que son los que están causando el problema para averiguar por qué suceden estos casos.

```python
def agrupaciones_inconsistentes(materias_por_semestre):
    """
    Permite filtrar las agrupaciones que tienen más de 9
    asignaturas por periodo y matricula.
    """
    # Obtenemos los registros que contiene inconsistencias
    inconsistencias = materias_por_semestre[materias_por_semestre["n_asignaturas"] > 9]
    # Cantidad de estudiantes que superan el limite permitido de asignaturas por semestre
    print("Agrupaciones con exceso de asignaturas por periodo: ", len(inconsistencias))
    # Ordenamos las inconsistencias de mayor a menor
    inconsistencias = inconsistencias.sort_values(by = ["n_matricula", "periodo", "n_asignaturas"], ascending = False)
    return(inconsistencias)
```

```python
# Calculamos las inconsistencias para el año actual
inconsistencias = agrupaciones_inconsistentes(materias_por_semestre)
# Mostramos esas inconsistencias
inconsistencias
```

Como pudimos ver anteriormente, tenemos 4 agrupaciones que nos arrojan 10 registros de asignaturas por semestre. En total se espera observar 40 registros. Filtramos los datos en el conjunto original por periodo y matricula usando como base los datos de las agrupaciones inconsistentes, para lograrlo ejecutamos un right join.

En el siguiente dataframe podemos observar los registros de un estudiante y encontramos que:

- En algunos casos aparecen dos materias repetidas dos veces en el mismo periodo pero con diferentes calificaciones finales, lo que indicaría que en el semestre se reprobó la asignatura y despues se presentó un examen extraordinario (que casualmente se registra bajo el mismo periodo).

```python
def registros_inconsistentes(dataframe, inconsistencias):
    """
    Permite obtener los registros que corresponden a las
    agrupaciones inconsistentes mostradas anteriormente.
    """
    # Mostramos los registros de las 4 agrupaciones anteriores
    return(pd.merge(dataframe, inconsistencias[["periodo", "n_matricula"]], on = ["periodo", "n_matricula"], how = "right").sort_values(by = ["n_matricula", "clave"], ascending = False))
```

```python
# Filtramos los registros inconsistentes para el año y los mostramos
registros_inconsistentes(datos["2016"], inconsistencias)
```

## Imputación de valores faltantes

Para esta sección es necesario explora los conjuntos de calificaciones en busca de valores nulos, habrá qué procesarlos, una estrategia sería usar alguna medida estadística como el promedio de las observaciones que sí se tienen e imputarlo en los registros faltantes

## Resumen de Dataframe en sus años correspondientes

A continuación, se hace un resumen de los valores nulos que hay en nuestros conjuntos de datos, más adelate se hará una exploración de estos valores y se tomarán decisiones. Como por ejemplo, imputar valores o simplemente eliminarlos.

```python
def resumen_valores_faltantes(dataframe):
    """
    Esta función permite obtener una tabla con un conteo de
    los valores faltantes representados con NaN
    en cada variable (columna)
    """
    return(dataframe.isnull().sum().reset_index(name = "valores_faltantes"))
```

### Eliminamos las cargas academicas sin datos en la variable "promediofinal"

```python
nan_rows = datos[datos["promediofinal"].isnull()] #matriculas con datos nulos
matriculas = nan_rows["n_matricula"].unique()
periodo = nan_rows["periodo"].unique()

#for a in matriculas:
#  nulos = len(nan_rows[nan_rows["n_matricula"]==a]) #Tamaño de datos nulos en la carga
#  total_asig = len(datos["2020_2021"][datos["2020_2021"]["n_matricula"]==a]) # Tamaño de la carga
#  print("matricula: ",a ," materias: ", total_asig, "nulos", nulos)

for per in periodo:
  print("=======================================================")
  for mat in matriculas:
    nulos = len(nan_rows[(nan_rows["n_matricula"]==mat) & (nan_rows["periodo"]== per)]) #Tamaño de datos nulos en la carga
    total_asig = len(datos[(datos["n_matricula"]==mat) & (datos["periodo"]==per)]) # Tamaño de la carga
    print("matricula: ",mat ,"periodo: ", per,"materias: ", total_asig, "nulos", nulos)

    if((nulos > 0) & (nulos == total_asig)): # En caso de que el tamaño de la carga sea igual al tamaño de los valores nulos entonces eliminamos
      print("mat ", mat, "periodo ",per)
      datos = datos.drop(datos[((datos["n_matricula"]==mat) & (datos["periodo"]==per))].index)
```
