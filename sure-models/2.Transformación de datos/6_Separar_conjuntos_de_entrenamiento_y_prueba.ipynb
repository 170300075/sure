## Estandarización

```r
est <- read.csv("predictores_train_VIF_EST.csv")
```

```r
nrow(est)
```

```r
#mean
mean(est$quintil_asig)

#sd
sd(est$quintil_asig)
```

```r
boxplot(est)
```

```r
stripchart(est$tasa_aprob_per_prev, method = "jitter", pch = 19,add = FALSE,col = "blue", vertical = TRUE)
```

```r
# calculamos el 70%
num_reg_entrenamiento<-as.integer(0.70*nrow(est))
num_reg_entrenamiento
```

```r
# Creamos una vector de 70% de los registros aleatorio
v_train <- sample(nrow(est), num_reg_entrenamiento)

# Creamos el conjunto de registros de entrenamiento pasando ese vector a la tabla 
train <- est[v_train,]
# Creamos los datos de comprobación o test (notese el -)
test <- est[-v_train , ]
```

```r
nrow(train)
```

```r
nrow(test)
```

```r
names(train)
```

===

```r
#Exportar datos
write.csv(train, "train_est.csv", row.names = FALSE)
write.csv(test, "test_est.csv", row.names = FALSE)
```
